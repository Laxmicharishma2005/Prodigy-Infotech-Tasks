```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
train_df = pd.read_csv('/content/Titanic-Dataset.csv')
test_df = pd.read_csv('/content/Titanic-Dataset.csv')

# Print the first few rows of the dataset
print(train_df.head())

# Check for missing values
print(train_df.isnull().sum())

# Drop the rows with missing values
train_df.dropna(inplace=True)

# Convert the 'Sex' column to a categorical variable
train_df['Sex'] = train_df['Sex'].map({'male': 0, 'female': 1})

# Convert the 'Embarked' column to a categorical variable
train_df['Embarked'] = train_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Plot the distribution of the 'Age' column
plt.figure(figsize=(10, 6))
sns.histplot(train_df['Age'], bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plot the distribution of the 'Fare' column
plt.figure(figsize=(10, 6))
sns.histplot(train_df['Fare'], bins=20)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()

# Plot the relationship between 'Age' and 'Fare'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=train_df)
plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# Plot the relationship between 'Sex' and 'Survived'
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', data=train_df)
plt.title('Relationship between Sex and Survived')
plt.xlabel('Sex')
plt.ylabel('Survived')
plt.show()

# Plot the relationship between 'Pclass' and 'Survived'
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', data=train_df)
plt.title('Relationship between Pclass and Survived')
plt.xlabel('Pclass')
plt.ylabel('Survived')
plt.show()
```
